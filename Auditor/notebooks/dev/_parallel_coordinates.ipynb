```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Feature 1': [1, 2, 3],
    'Feature 2': [4, 5, 6],
    'Feature 3': [7, 8, 9],
    'Class': ['A', 'B', 'A']
}

df = pd.DataFrame(data)

# Normalize data (optional)
df_normalized = df.iloc[:, :-1].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

# Add the class column back
df_normalized['Class'] = df['Class']

# Create the plot
plt.figure(figsize=(10, 5))
for i, row in df_normalized.iterrows():
    plt.plot(df_normalized.columns[:-1], row[:-1], label=row['Class'], alpha=0.5)

# Add labels and title
plt.title('Parallel Coordinate Plot')
plt.xlabel('Features')
plt.ylabel('Normalized Value')
plt.legend(loc='best')
plt.grid()

# Show the plot
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

# Create the plot
parallel_coordinates(df, 'Class', color=['blue', 'red'])

# Show the plot
plt.show()
```
